```python
from transformers import AutoTokenizer
from datasets import load_dataset

checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
raw_datasets = load_dataset("wikitext", "wikitext-2-raw-v1")
```

```python
sequence = raw_datasets["train"][10]["text"]
```

```python
tokenized_sequence = tokenizer.tokenize(sequence)
```

```python
tokenized_sequence[5] = "[MASK]"
```

```python
masked_sequence = tokenizer.convert_tokens_to_string(tokenized_sequence)
```

```python
from transformers import pipeline 

unmasker = pipeline('fill-mask', model='distilbert-base-cased')
unmasker(masked_sequence)
```
