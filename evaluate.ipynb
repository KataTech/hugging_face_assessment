```python
import evaluate

predictions = ["Evaluate is a library to evaluate Machine Learning models"]
references = ["Evaluate is a library to evaluate ML models"]
bleu = evaluate.load("bleu")
results = bleu.compute(predictions=predictions, references=references)
print(results)
```
